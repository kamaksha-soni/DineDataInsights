Calculate the average length of reviews and
explore if there is a relationship between
review length and rating.

```python
# Drop rows with missing values DATA CLEANING:
data = data.dropna(subset=['Votes', 'Aggregate rating'])

# Convert Votes to numeric 
data['Votes'] = pd.to_numeric(data['Votes'], errors='coerce')

# Calculate the average number of votes 
average_votes = data['Votes'].mean()
print(f"The average number of votes (proxy for review length) is: {average_votes:.2f}")

# relationship between votes (review length) and aggregate rating
rating_votes = data.groupby('Aggregate rating')['Votes'].mean()

print("\nAverage Votes by Aggregate Rating:")
print(rating_votes)

fig = px.bar(
    rating_votes,
    x=rating_votes.index,
    y='Votes',
    labels={'Votes': 'Average Votes (Proxy for Review Length)', 'Aggregate rating': 'Aggregate Rating'},
    title='Relationship Between Review Length (Votes) and Rating'
)
fig.show()
```

Task: Votes Analysis

Identify the restaurants with the highest and
lowest number of votes.

```python
# DATA Cleaning:
data['Votes'] = pd.to_numeric(data['Votes'], errors='coerce')

# Drop rows with missing values:
data = data.dropna(subset=['Votes'])

# restaurant with the highest number of votes
highest_votes = data.loc[data['Votes'].idxmax()]

# Frestaurant with the lowest number of votes:
lowest_votes = data.loc[data['Votes'].idxmin()]

# Display results:
print("Restaurant with the highest number of votes:")
print(highest_votes[['Votes']])

print("\nRestaurant with the lowest number of votes:")
print(lowest_votes[[ 'Votes']])
```

Analyze if there is a correlation between the
number of votes and the rating of a
restaurant.

```python
# DATA Cleaning:
data['Votes'] = pd.to_numeric(data['Votes'], errors='coerce')
data['Aggregate rating'] = pd.to_numeric(data['Aggregate rating'], errors='coerce')

# Drop rows with missing or zero values in Votes and Aggregate rating:
data = data.dropna(subset=['Votes', 'Aggregate rating'])
data = data[(data['Votes'] > 0) & (data['Aggregate rating'] > 0)]

# Calculate correlation coefficient:
correlation = np.corrcoef(data['Votes'], data['Aggregate rating'])[0, 1]
print(f"Correlation between Votes and Aggregate Rating: {correlation:.2f}")

# Create Scatter Plot:
fig = px.scatter(
    data, 
    x='Votes', 
    y='Aggregate rating', 
    title='Relationship Between Votes and Aggregate Rating',
    labels={'Votes': 'Number of Votes', 'Aggregate rating': 'Aggregate Rating'},
    opacity=0.8
)

# Show the plot
fig.show()

```

Task: Price Range vs. Online Delivery and
Table Booking

```python
# DATA Cleaning:
data['Price range'] = pd.to_numeric(data['Price range'], errors='coerce')
data['Has Online delivery'] = data['Has Online delivery'].str.strip().str.capitalize() 
data['Has Table booking'] = data['Has Table booking'].str.strip().str.capitalize()  

# Groupby DATA:
price_delivery_table = data.groupby(['Price range', 'Has Online delivery', 'Has Table booking']).size().reset_index(name='Count')
total_counts = price_delivery_table.groupby('Price range')['Count'].transform('sum')
price_delivery_table['Percentage'] = (price_delivery_table['Count'] / total_counts) * 100

# Bar Chart Representation of data to visualize clearly:
fig = px.bar(
    price_delivery_table,
    x='Price range',
    y='Percentage',
    color='Has Online delivery',
    text='Percentage',
    title='Price Range v/s Online Delivery and Table Booking',
    labels={
        'Price range': 'Price Range',
        'Percentage': 'Percentage of Restaurants',
        'Has Online delivery': 'Online Delivery'
    }
)
fig.update_traces(texttemplate='%{text:.2f}%', textposition='inside')
fig.show()
```

Analyze if there is a relationship between the
price range and the availability of online
delivery and table booking.


```python
# DATA Cleaning:
data['Price range'] = pd.to_numeric(data['Price range'], errors='coerce')
data['Has Online delivery'] = data['Has Online delivery'].str.strip().str.capitalize()
data['Has Table booking'] = data['Has Table booking'].str.strip().str.capitalize()

# Group by DATA by : Price Range, Online Delivery, and Table Booking:
price_range_analysis = (
    data.groupby(['Price range', 'Has Online delivery', 'Has Table booking']).size().reset_index(name='Count')
)

# Calc Total COUNTS for Price Ranage:
total_counts = price_range_analysis.groupby('Price range')['Count'].transform('sum')

# Calculate percentages:
price_range_analysis['Percentage'] = (price_range_analysis['Count'] / total_counts) * 100

print(price_range_analysis)
```

Determine if higher-priced restaurants are
more likely to offer these services.

```python
#Group Data by Price Range and Count the Services:
Price_range = data.groupby('Price range').agg(
    online_delivery_count=('Has Online delivery', lambda x: (x == 'Yes').sum()),
    table_booking_count=('Has Table booking', lambda x: (x == 'Yes').sum()),
    total_restaurants=('Has Online delivery', 'count')
).reset_index()

# Calculate Percentages
Price_range['Online Delivery %'] = (Price_range['online_delivery_count'] / Price_range['total_restaurants']) * 100
Price_range['Table Booking %'] = (Price_range['table_booking_count'] / Price_range['total_restaurants']) * 100

print(Price_range)
```
